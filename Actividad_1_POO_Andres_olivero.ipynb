Ejercicio resuelto numero 4

```python
class Edades:
    def __init__(self, EDJUAN):
        self.EDJUAN = EDJUAN
        self.EDALBER = 0
        self.EDANA = 0
        self.EDMAMA = 0

    def calcular(self):
        self.EDALBER = 2 * self.EDJUAN / 3
        self.EDANA = 4 * self.EDJUAN / 3
        self.EDMAMA = self.EDJUAN + self.EDALBER + self.EDANA

    def mostrar(self):
        print("LAS EDADES SON:")
        print("ALBERTO =", self.EDALBER)
        print("JUAN =", self.EDJUAN)
        print("ANA =", self.EDANA)
        print("MAMA =", self.EDMAMA)



EDJUAN = float(input("Ingrese la edad de Juan: "))

edades = Edades(EDJUAN)
edades.calcular()
edades.mostrar()

```

Ejercicio propuesto 12

```python
class Empleado:
    def __init__(self):
        self.HORAS = 48
        self.VALOR_HORA = 5000
        self.SAL_BRUTO = 0
        self.RETENCION = 0
        self.SAL_NETO = 0

    def calcular_salario(self):
        self.SAL_BRUTO = self.HORAS * self.VALOR_HORA
        self.RETENCION = self.SAL_BRUTO * 0.125
        self.SAL_NETO = self.SAL_BRUTO - self.RETENCION

    def mostrar_resultados(self):
        print("SALARIO BRUTO:", self.SAL_BRUTO)
        print("RETENCION EN LA FUENTE:", self.RETENCION)
        print("SALARIO NETO:", self.SAL_NETO)


# INICIO
empleado = Empleado()
empleado.calcular_salario()
empleado.mostrar_resultados()
# FIN
```

ejercicio propuesto 14


```python
class Potencias:
    def __init__(self, NUM):
        self.NUM = NUM
        self.CUADRADO = 0
        self.CUBO = 0

    def calcular(self):
        self.CUADRADO = self.NUM * self.NUM
        self.CUBO = self.NUM * self.NUM * self.NUM

    def mostrar(self):
        print("EL CUADRADO ES:", self.CUADRADO)
        print("EL CUBO ES:", self.CUBO)



NUM = float(input("Ingrese un número: "))

potencia = Potencias(NUM)
potencia.calcular()
potencia.mostrar()

```

```python
import math

class Circulo:
    def __init__(self, RADIO):
        self.RADIO = RADIO
        self.AREA = 0
        self.LONGITUD = 0

    def calcular(self):
        self.AREA = math.pi * self.RADIO * self.RADIO
        self.LONGITUD = 2 * math.pi * self.RADIO

    def mostrar(self):
        print("EL AREA DEL CIRCULO ES:", format(self.AREA, ".2f"))
        print("LA LONGITUD DE LA CIRCUNFERENCIA ES:", format(self.LONGITUD, ".2f"))



RADIO = float(input("Ingrese el radio del círculo: "))

circulo = Circulo(RADIO)
circulo.calcular()
circulo.mostrar()

```
